```python
import nltk

nltk.download('popular')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('words')
nltk.download('maxent_ne_chunker')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
import re

lemmatizer=WordNetLemmatizer()
tokenizer=RegexpTokenizer(r'\w+')
stopwords_=set(stopwords.words('english'))
stemmer= PorterStemmer()
```

```python
def preprocess(text):

    text = text if type(text)!= float else ''
    text = text if len(text)!=0 else ''
    # Removing HTML tags
    clean = re.compile('<.*?>')
    text = re.sub(clean, '', text)
    # Removing tags
    text = re.sub(r'[^\w]', ' ', text)
    # remove special characters and digits
    text = re.sub("(\\d|\\W)+", " ", text)
    text = re.sub(r"\b[a-zA-Z]\b", "", text)
    # Lowercase
    text = text.lower()

    # removing white space
    text.strip()
    # Tokenization, Stopwords removal, Lemmatization
    return " ".join([lemmatizer.lemmatize(word)
                     for word in [i for i in tokenizer.tokenize(text.lower())
                                  if not i in stopwords_]])
```

```python
def tag_attributes(x):
    import nltk
    nltk.download('stopwords')
    from nltk.corpus import stopwords
    from collections import OrderedDict
    stop_words = set(stopwords.words('english'))
    x = preprocess(x)
    x = x.split()
    x = [word for word in x if word not in stop_words]
    # x = [str(word) for word in x if type(word)== int or len(word)>=2]
    x = list(OrderedDict.fromkeys(x))
    return x
```

```python
################## Product Color Fetching ##################
def get_product_color(img_url):
    def remove_background_from_url(url):
        import requests
        import io
        from io import BytesIO
        from PIL import Image
        from rembg import remove
        import base64
        if len(url)<=0:
            return None
        else:
            response = requests.get(url)
            try:
                input_image = Image.open(BytesIO(response.content))
                output_image = remove(input_image)
                output_buffer = BytesIO()
                output_image.save(output_buffer, format="PNG")
                output_image_data = output_buffer.getvalue()
                base64_image = base64.b64encode(output_image_data).decode('utf-8')
                image_bytes = io.BytesIO(base64.b64decode(base64_image))
                image = Image.open(image_bytes)
            except Exception as e:
                # image = image_bytes = None
                image_bytes = None
            return image_bytes

    def closest_colour(requested_colour):
        import webcolors
        min_colours = {}
        for key, name in webcolors.CSS3_HEX_TO_NAMES.items():
            r_c, g_c, b_c = webcolors.hex_to_rgb(key)
            rd = (r_c - requested_colour[0]) ** 2
            gd = (g_c - requested_colour[1]) ** 2
            bd = (b_c - requested_colour[2]) ** 2
            min_colours[(rd + gd + bd)] = name
        return min_colours[min(min_colours.keys())]

    def get_colour_name(requested_colour):
        try:
            import webcolors
            # closest_name = actual_name = webcolors.rgb_to_name(requested_colour)
            closest_name = webcolors.rgb_to_name(requested_colour)
        except ValueError:
            closest_name = closest_colour(requested_colour)
            # actual_name = None
        # return actual_name, closest_name
        return closest_name

    image_bytes = remove_background_from_url(img_url)

    if image_bytes!=None:
        # display(image)
        from Pylette import extract_colors
        size = 3 # Size 3 is fixed, select the patelle[1](second color) as the product color
        palette = extract_colors(image=image_bytes, palette_size=size, resize=True)
        # palette.display(save_to_file=False)

        requested_colour = palette[1].rgb
        closest_color = get_colour_name(requested_colour)
        # print("Actual colour name:", actual_color, ", closest colour name:", closest_color)
        # print(closest_color,'\n')
        return closest_color
    else:
        return "None"
```

```python
def generate_tags_data(df1):
    productVarData = df1

    productVarData.fillna('', inplace=True)

    productVarData['NER_Text'] = productVarData["ProductName"] + ' ' + productVarData["Category"] + ' ' + productVarData['Brand'] + ' ' + productVarData['TagAttributes'] + ' ' + productVarData["VariantName"] + ' ' + productVarData["Color"] + ' ' + productVarData["Size"]+ ' ' + productVarData["ProductCode"]+ ' ' + productVarData["ProductVariantCode"]
    productVarData['NER_Text'] = productVarData['NER_Text'].apply(lambda x: preprocess(x))

    productVarData['Product_Tags'] = productVarData['NER_Text'].apply(lambda x: tag_attributes(x))
    if productVarData["Color"].isna().any() or (productVarData["Color"].str.strip() == "").any():
        productVarData['ProductColor'] = productVarData['ImageURL'].apply(lambda url: get_product_color(str(url)))
        productVarData['ProductColor'] = productVarData['ProductColor'].apply(lambda x: [] if x == "None" else [x])
        productVarData['Product_Tags'] = productVarData['Product_Tags'] + productVarData['ProductColor']
        productVarData = productVarData.drop(columns=['NER_Text', 'ImageURL', 'ProductColor'], axis=1)
    else:
        productVarData = productVarData.drop(columns=['NER_Text', 'ImageURL'], axis=1)

    productVarData['Product_Tags'] = productVarData['Product_Tags'].apply(lambda x: json.dumps(x))



    return productVarData

```
